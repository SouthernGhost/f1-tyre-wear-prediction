#### Plot lap times for a selected driver, compund, grand prix and year and calculate correlation between
#### lap times and lap number

```python
def plot_laptimes(driver,comp,gp,year):
    try:
        x = df[(df['Driver']==driver) 
            & (df['GrandPrix']==gp) 
            & (df['Year']==year)
            & (df['Compound']==comp)][['LapTime','LapNumber']]
        
        x.plot(x='LapNumber', kind='line',
                title=f'Lap Times for {driver}',
                xticks=(range(int(x['LapNumber'].min()),int(x['LapNumber'].max()),1)),
                grid=True,
                rot=30,
                figsize=(18,6))
        
        _laptimes = x['LapTime']
        string = 'LapNumber'
        print(f'Correlation between lap time and lap number of {driver}: {x[string].corr(_laptimes):.2f}')
    except:
        print(f'Data for {driver} not found.')

drv_drop = widgets.Dropdown(options=df.sort_values(by='Driver',ascending=True)['Driver'].unique(), 
                                    description='Driver: ')

comp_drop = widgets.Dropdown(options=df['Compound'].unique(),
                                    description='Compound: ')

gp_drop = widgets.Dropdown(options=df.sort_values(by=['RoundNumber','Year'])['GrandPrix'].unique(),
                                    description='Grand Prix: ')

year_drop = widgets.Dropdown(options=df['Year'].unique(),
                                    description='Year: ')

widgets.interactive(plot_laptimes, driver=drv_drop, comp=comp_drop, gp=gp_drop, year=year_drop)
```
